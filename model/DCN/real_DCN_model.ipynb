```python
ratings_dataset = tf.data.Dataset.from_tensor_slices({
    'user_id': ratings['user_id'].values,
    'song_name': ratings['song_name'].values,
    'user_rating': ratings['user_rating'].values,
    'sentiment_score' : ratings['sentiment_score'].values,
    'age' : ratings['age'].values,
    'gender' : ratings['age'].values,
})

new_song_dataset = tf.data.Dataset.from_tensor_slices({
    'song_name': ratings['song_name'].values,  
})
```

```python
# timestamp + 감성점수 + 나이 + 성별
class UserModel(tf.keras.Model):

    def __init__(self, use_timestamps):
        super().__init__()

        self._use_timestamps = use_timestamps

        # User id embedding
        self.user_embedding = tf.keras.Sequential([
            tf.keras.layers.StringLookup(
                vocabulary=unique_user_ids,
                mask_token=None),
            tf.keras.layers.Embedding(len(unique_user_ids) + 1, 32)
        ])


        # Sentiment score embedding
        self.sentiment_score_embedding = tf.keras.Sequential([
            tf.keras.layers.Embedding(6, 16)  
        ])

        # Age embedding
        self.age_embedding = tf.keras.Sequential([
            tf.keras.layers.Embedding(7, 16)  
        ])

        # Gender embedding
        self.gender_embedding = tf.keras.Sequential([
            tf.keras.layers.Embedding(10, 16)  
        ])

    def call(self, inputs):
        if not self._use_timestamps:
            return self.user_embedding(inputs['user_id'])

        return tf.concat([
            self.user_embedding(inputs['user_id']),
            self.sentiment_score_embedding(inputs['sentiment_score']),
            self.age_embedding(inputs['age']),
            self.gender_embedding(inputs['gender']),
        ], axis=1)
```

```python
class SongModel(tf.keras.Model):

    def __init__(self, use_title_text=True,output_dim=80):
        super().__init__()
        max_tokens=1_000
        self.use_title_text = use_title_text

        self.title_embedding = tf.keras.Sequential([
            tf.keras.layers.StringLookup(
                vocabulary=unique_song_titles, mask_token=None),
            tf.keras.layers.Embedding(len(unique_song_titles)+1, 32)
        ])

        if use_title_text:
            self.title_vectorizer = tf.keras.layers.TextVectorization(
                max_tokens=max_tokens)
            self.title_vectorizer.adapt(new_song_dataset.map(lambda x: x['song_name']))

            self.title_text_embedding = tf.keras.Sequential([
                self.title_vectorizer,
                tf.keras.layers.Embedding(max_tokens, 32, mask_zero=True),
                tf.keras.layers.GlobalAveragePooling1D()
            ])
        
    def call(self, inputs):
        if not self.use_title_text:
            return self.title_embedding(inputs['song_name'])
        
        title_embedding = self.title_embedding(inputs['song_name'])
        title_text_embedding = self.title_text_embedding(inputs['song_name'])
        return tf.concat([title_embedding, title_text_embedding], axis=1)
```

```python
class MovielensModel(tfrs.models.Model):
    
    def __init__(self, use_timestamps=False, use_title_text=True):
        super().__init__()
        
        self.rating_weight = 0.5
        self.retrieval_weight = 0.5

        # 총 고유 'user_id' 값의 개수를 계산합니다.
        num_unique_users = ratings['user_id'].nunique() + 1
        
        # 사용자 모델의 임베딩 레이어 크기를 num_unique_users로 설정합니다.
        self.user_model = tf.keras.Sequential([
            tf.keras.layers.Embedding(num_unique_users, 32),
            tf.keras.layers.Dense(32, activation="relu"),
        ])
        
        self.song_model = tf.keras.Sequential([
            SongModel(use_title_text),
            tf.keras.layers.Dense(80)
        ])
        self.sentiment_embedding = tf.keras.layers.Embedding(6, 16)
        self.age_embedding = tf.keras.layers.Embedding(7, 16)
        self.gender_embedding = tf.keras.layers.Embedding(10, 16)

        # Ranking model with DCN using keras Functional API
        x0 = tf.keras.Input(shape=(160,))
        x1 = tfrs.layers.dcn.Cross()(x0, x0)
        x2 = tfrs.layers.dcn.Cross()(x0, x1)
        d1 = tf.keras.layers.Dense(256, activation='relu')(x2)
        d2 = tf.keras.layers.Dense(64, activation='relu')(d1)
        output = tf.keras.layers.Dense(1)(d2)
        self.rating_model = tf.keras.Model(inputs=x0, outputs=output)
    
        # Multi-tasks
        self.rating_task: tf.keras.layers.Layer = tfrs.tasks.Ranking(
            loss=tf.keras.losses.MeanSquaredError(),
            metrics=[tf.keras.metrics.RootMeanSquaredError()]
        )
        self.retrieval_task: tf.keras.layers.Layer = tfrs.tasks.Retrieval(
            metrics=tfrs.metrics.FactorizedTopK(
                candidates=new_song_dataset.batch(128).map(self.song_model)
            )
        )
    def call(self, features: Dict[Text, tf.Tensor]) -> tf.Tensor:
        user_embeddings = self.user_model(features['user_id'])
        sentiment_embed = self.sentiment_embedding(features['sentiment_score'])
        age_embed = self.age_embedding(features['age'])
        gender_embed = self.gender_embedding(features['gender'])
        user_embeddings = tf.concat([user_embeddings, sentiment_embed, age_embed, gender_embed], axis=1)
        
        # user_embeddings = self.user_model(features['user_id'])
        song_embeddings = self.song_model({'song_name': features['song_name']})

        return (
            user_embeddings, 
            song_embeddings,
            self.rating_model(tf.concat([user_embeddings, song_embeddings], axis=1))
        )        

    def compute_loss(self, features: Dict[Text, tf.Tensor], training=False) -> tf.Tensor:
        user_embeddings, song_embeddings, rating_predictions = self(features)

        # 'song_embeddings'를 'tfrs.metrics.FactorizedTopK' 메트릭스에 전달
        retrieval_loss = self.retrieval_task(user_embeddings, song_embeddings)
        rating_loss = self.rating_task(
            labels=features['user_rating'],
            predictions=rating_predictions
        )

        return (self.rating_weight * rating_loss + self.retrieval_weight * retrieval_loss)

```

```python
tf.random.set_seed(7)
shuffled = ratings_dataset.shuffle(9_000, seed=7,
                reshuffle_each_iteration=False)

train = shuffled.take(7_200)
test = shuffled.skip(7_200).take(1_800)

cached_train = train.shuffle(7_200).batch(50).cache()
cached_test = test.batch(10).cache()
```

```python
model = MovielensModel(use_timestamps=True, use_title_text=True)
model.compile(optimizer=tf.keras.optimizers.Adagrad(0.1))

model.fit(cached_train, epochs=3,steps_per_epoch=10)#steps_per_epoch=5

train_acc = model.evaluate(
    cached_train, return_dict=True)['factorized_top_k/top_100_categorical_accuracy']
test_acc = model.evaluate(
    cached_test, return_dict=True)['factorized_top_k/top_100_categorical_accuracy']

print(f'Top-100 accuracy (train): {train_acc:.2f}')
print(f'Top-100 accuracy (test): {test_acc:.2f}')
```
